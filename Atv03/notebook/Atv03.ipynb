Usando a classe JOptionPane para entrada de dados, receba a nota de duas provas e de um trabalho. 

```java
import javax.swing.JOptionPane;
String prova1 = JOptionPane.showInputDialog("Digite a nota da 1ª prova: ");
double nota1 = Double.parseDouble(prova1);
String prova2 = JOptionPane.showInputDialog("Digite a nota da 2ª prova:");
double nota2 = Double.parseDouble(prova2);
String trabalho = JOptionPane.showInputDialog("Digite a nota do trabalho:");
double notaTrabalho = Double.parseDouble(trabalho);
```

Calcule a média 

```java
double media = (nota1 + nota2 + notaTrabalho) / 3;
```

Mostre a média usando o showMessageDialog da classe JOptionPane

```java
JOptionPane.showMessageDialog(null, "A média é: " + media);
```
