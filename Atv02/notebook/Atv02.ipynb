Demonstrar a execução do cálculo e utilizar um comentário para afirmar que é um programa de exemplo de folha de pagamento.

```java
import java.util.Scanner;
int b, c, a = 0;
Scanner s = new Scanner(System.in);
b = s.nextInt();
c = s.nextInt();
a = b * c;
System.out.println("O produto de b e c é: " + a);
// Programa de exemplo de folha de pagamento.
```
